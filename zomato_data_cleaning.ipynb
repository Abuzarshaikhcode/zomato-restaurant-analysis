#### Cleaning Data 

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Cuisine'])
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['Rating'] = df['Rating'].astype(str).str.extract('(\d+\.\d+)')
df['Rating'] = df['Rating'].astype(float)
```

```python
df
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean()).round(1)
```

```python
df['Average Price'] = df['Average Price'].astype(str).str.extract('(\d+)')
df['Average Price'] = df['Average Price'].astype(float)
```

```python
df['Average Delivery Time'] = df['Average Delivery Time'].astype(str).str.extract('(\d+)')
df['Average Delivery Time'] = df['Average Delivery Time'].astype(float)
```

```python
print(df[['Average Price', 'Average Delivery Time']].isnull().sum())
```

```python
cols = ['Average Price', 'Average Delivery Time']
df[cols] = df[cols].fillna(0).astype(int)
```

```python
df = df[(df['Average Price'] != 0) & (df['Average Delivery Time'] != 0)]
```

```python
df.describe()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.to_csv("clean_zomato.csv", index=False)
```
